**RAG workflow**: The system will chunk text documents, create embeddings, stores them in a vector database, and uses them to enhance LLM responses.

When querying, the vector database compares the indexed vectors to the query vector to determine the nearest vector neighbor.

We can use one of the following similarity measures to find the nearest neighbor:
1. Cosine similarity (focuses on direction, ignore magnitude)
2. Euclidian distance (captures magnitude difference well)
3. Dot product (combines magnitude and direction effectively)
